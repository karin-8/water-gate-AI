```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Data
sections = ["Section A", "Section B", "Section C"]
widths = [120, 110, 40]         # in cm
water_levels = [12, 8, 6]       # in m
gate_heights = [2, 8, 1]        # in cm
gate_positions = [0, 1, 2]

# Plotting
fig, ax = plt.subplots(figsize=(12, 5))
# Set bar width to fill the space entirely
bar_width = 1.0
# Draw water level bars
bars = ax.bar(gate_positions, water_levels, width=bar_width, color='deepskyblue', align='edge')

# Add max/min lines
ax.axhline(10, color='red', linestyle='-', label='Max')
ax.axhline(6, color='orange', linestyle='-', label='Min')

# Add gates
for i, height in zip(gate_positions, gate_heights):
    gate = Rectangle((i - 0.05, 0), 0.1, height/5,  # Convert cm to m
                     linewidth=2, edgecolor=None, facecolor='white')
    gatewall = Rectangle((i - 0.05, height/5), 0.1, 10 - height/5,  # Convert cm to m
                 linewidth=2, edgecolor='black', facecolor='grey')
    ax.add_patch(gate)
    ax.add_patch(gatewall)
    ax.text(i, height / 100 + 0.2, f"{int(height)} cm: {height/20*100}%", ha='center', fontsize=9)

# Annotate water height
for i, h in zip(gate_positions, water_levels):
    ax.text(i+0.5, h - 0.5, f"+{h} m", ha='center', color='white', fontsize=10)

# X labels
ax.set_xticks(gate_positions)
ax.set_xticklabels([f"{w} cms" for w in widths])

# Aesthetic
ax.set_ylim(0, 14)
ax.set_ylabel("Height (m)")
ax.set_title("Water Levels and Gate Openings")
ax.legend()

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Data
sections = ["Section A", "Section B", "Section C"]
widths = [120, 110, 40]         # in cm
water_levels = [12, 8, 6]       # in m
gate_heights = [2, 8, 1]        # in cm
gate_positions = [0, 1, 2]

# Plotting
fig, ax = plt.subplots(figsize=(12, 5))
bar_width = 1.0
bars = ax.bar(gate_positions, water_levels, width=bar_width, color='deepskyblue', align='edge')

# Max/Min lines
ax.axhline(10, color='red', linestyle='-', label='Max')
ax.axhline(6, color='orange', linestyle='-', label='Min')

# Add gates and callouts
for i, height in zip(gate_positions, gate_heights):
    gate_height_m = height / 100 * 20 # cm to m
    gate_open_ratio = height / 20 * 100  # max = 20 cm

    gate = Rectangle((i - 0.05, 0), 0.1, gate_height_m, facecolor='deepskyblue')
    gatewall = Rectangle((i - 0.05, gate_height_m), 0.1, 10 - gate_height_m, edgecolor='black', facecolor='grey')
    ax.add_patch(gate)
    ax.add_patch(gatewall)

    # Add annotation (callout)
    ax.annotate(f"{height} cm\n{gate_open_ratio:.0f}%",
                xy=(i, gate_height_m),
                xytext=(i + 0.2, gate_height_m + 2),
                arrowprops=dict(facecolor='black', arrowstyle='->'),
                ha='right', fontsize=9, bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow", ec="black", lw=0.5))

# Water level text
for i, h in zip(gate_positions, water_levels):
    ax.text(i + 0.5, h - 0.5, f"+{h} m", ha='center', color='white', fontsize=10)



# X labels
ax.set_xticks(gate_positions)
ax.set_xticklabels([f"{w} cms" for w in widths])

# Aesthetic
ax.set_ylim(0, 14)
ax.set_ylabel("Height (m)")
ax.set_title("Water Levels and Gate Openings")
ax.legend()

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

max_w_height = 15
min_w_height = 6

# Data
sections = ["Section A", "Section B", "Section C"]
qs = [120, 120, 40, 30, 20]         # in cms
water_levels = [12, 8, 6, 2, 3]       # in m
gate_heights = [2, 8, 3, 4, 2]        # in cm
gate_positions = [0, 1, 2, 3, 4]
gate_names = ["C2", "Manorom", "Chong Khae", "Khok Krathiam", "Roeng Rang"]

# Plotting
fig, ax = plt.subplots(figsize=(12, 5))
bar_width = 1.0
bars = ax.bar(gate_positions, water_levels, width=bar_width, color='deepskyblue', align='edge')

# Max/Min lines
ax.axhline(max_w_height, color='red', linestyle='-', label='Max')
ax.axhline(min_w_height, color='orange', linestyle='-', label='Min')

# Add gates and callouts
for i, height in zip(gate_positions, gate_heights):
    if i != 0:
        gate_height_m = height / 100 * 20 # cm to m
        gate_open_ratio = height / 20 * 100  # max = 20 cm
    
        gate = Rectangle((i - 0.05, 0), 0.1, gate_height_m, facecolor='deepskyblue')
        gatewall = Rectangle((i - 0.05, gate_height_m), 0.1, max_w_height - gate_height_m, edgecolor='black', facecolor='grey')
        ax.add_patch(gate)
        ax.add_patch(gatewall)
    
        # Callout
        ax.annotate(f"{height} cm\n{gate_open_ratio:.0f}%",
                    xy=(i, gate_height_m),
                    xytext=(i + 0.2, gate_height_m + 2),
                    arrowprops=dict(facecolor='black', arrowstyle='->'),
                    ha='right', fontsize=9,
                    bbox=dict(boxstyle="round,pad=0.3", fc="lightyellow", ec="black", lw=0.5))
    
        # White dotted arrow inside the gate opening
        arrow_y = gate_height_m / 2
        ax.annotate("",
                    xy=(i + 0.08, arrow_y), xytext=(i - 0.06, arrow_y),
                    arrowprops=dict(arrowstyle="->", linestyle="-", color='white', lw=2))

# Water level text
for i, h in zip(gate_positions, water_levels):
    ax.text(i + 0.5, h - 0.5, f"+{h} m", ha='center', color='white', fontsize=10)

# X labels
ax.set_xticks(gate_positions)
ax.set_xticklabels([f"{gate_names[i]}\nQ = {qs[i]} cms" for i in range(len(qs))])

# Aesthetic
ax.set_ylim(0, 16)
ax.set_ylabel("Height (m)")
ax.set_title("Water Levels and Gate Openings")
ax.legend()

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
